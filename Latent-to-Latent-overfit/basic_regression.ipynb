```python
import torch
from torch import nn
from torch import optim
from torch.nn import functional as F
import numpy as np
```

```python
x = np.load('/home/ubuntu/Latent-Transfer/Latent-to-Latent-overfit/x.npy')
y = np.load('/home/ubuntu/Latent-Transfer/Latent-to-Latent-overfit/y.npy')

x = torch.tensor(x).cuda()
y = torch.tensor(y).cuda()
```

```python
x[:10], y[:10]
```

```python
x.max(), x.min(), y.max(), y.min()
```

```python
class Network(nn.Module):
    def __init__(self):
        super(Network, self).__init__()
        self.layer1 = nn.Linear(256, 256)
        self.dropout1 = nn.Dropout(0.25)
        self.layer2 = nn.Linear(256, 256)
    
    def forward(self, x):
        x = self.layer1(x)
        x = F.relu(x)
        x = self.dropout1(x)
        x = self.layer2(x)
        
        return x
```

```python
model = Network().cuda()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
```

```python
for _ in range(100):
    optimizer.zero_grad()
    output = model(x)
    loss = F.mse_loss(y, output)
    loss.backward()
    optimizer.step()
    print(loss)
```
